## __Scatter X and Y

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.array([0, 1, 2, 3, 4, 5, 6, 7.])
ys = np.array([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37])
```

```python
fig, ax = plt.subplots()
plt.xlabel("x")
plt.ylabel("y")
_ = ax.scatter(xs, ys)
```
